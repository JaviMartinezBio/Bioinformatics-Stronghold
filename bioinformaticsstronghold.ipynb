```python
def WascaliWabbit(months, offspring):
    if months == 1 or months == 2:
        return 1
    else:
        return WascaliWabbit(months - 1, offspring) + offspring * WascaliWabbit(months - 2, offspring)


WascaliWabbit(32, 3)
```
